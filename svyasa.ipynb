# Line SubPlot of all the signals

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract the channel data (columns excluding 'sec' and 'Vertex Reference')
channel_data =baseline.drop(columns=['sec', 'Vertex Reference'])

# Create subplots for each channel
fig, axs = plt.subplots(len(channel_data.columns), 1, figsize=(12, len(channel_data.columns) * 0.6), sharex=True)

# Plot each channel's data
for i, channel in enumerate(channel_data.columns):
    timestamps = baseline['sec']
    signal_values = channel_data[channel]
    
    axs[i].plot(timestamps, signal_values, color='blue', linewidth=1.5)
    axs[i].set_ylabel('Signal Amplitude')
    axs[i].set_title(channel)
    axs[i].grid(True)

# Set x-axis label and title for the last subplot
axs[-1].set_xlabel('Time (seconds)')
fig.suptitle('EEG Signal Line Plots')

plt.tight_layout()  # Adjust the spacing between subplots
plt.show()
```
